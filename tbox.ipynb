# SDM_3

```python
# import libraries
from rdflib import Graph,Literal,RDF,URIRef
from rdflib.namespace import FOAF,XSD,RDFS
from rdflib import Namespace
```

```python
g = Graph()
```

```python
# name space is used for the graph general URIref, now we can add 
# multiple objects with kg_sdm uri (/Author, /School) etc.
KG_SDM = Namespace("http://kg_sdm.org/")
g.bind("kg_sdm",KG_SDM)
```

## Properties
* It will automaticly add the Classes if inference is on

```python
# ###  http://www.gra.fo/schema/untitled-ekg#comment
# gf:comment rdf:type owl:ObjectProperty ;
#            rdfs:domain gf:Review_process ;
#            rdfs:range xsd:Xsdstring_d9cd77c5-ded2-48d4-ae88-51c27d887334 ;
#            rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#            rdfs:label "comment" .


g.add((KG_SDM.comment, RDFS.domain, KG_SDM.ReviewProcess))
g.add((KG_SDM.comment, RDFS.range, XSD.string))
g.add((KG_SDM.comment, RDFS.label, Literal("comment")))



# ###  http://www.gra.fo/schema/untitled-ekg#goes_through
# gf:goes_through rdf:type owl:ObjectProperty ;
#                 rdfs:domain gf:Submission ;
#                 rdfs:range gf:Review_process ;
#                 rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#                 rdfs:label ":goes_through" .


g.add((KG_SDM.goes_through, RDFS.domain, KG_SDM.Submission))
g.add((KG_SDM.goes_through, RDFS.range, KG_SDM.ReviewProcess))
g.add((KG_SDM.goes_through, RDFS.label, Literal("goes_through")))




# ###  http://www.gra.fo/schema/untitled-ekg#handles
# gf:handles rdf:type owl:ObjectProperty ;
#            rdfs:domain gf:Handers ;
#            rdfs:range gf:Venue ;
#            rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#            rdfs:label "handles" .




g.add((KG_SDM.handles, RDFS.domain, KG_SDM.Handler))
g.add((KG_SDM.handles, RDFS.range, KG_SDM.Venue))
g.add((KG_SDM.handles, RDFS.label, Literal("handles")))





# ###  http://www.gra.fo/schema/untitled-ekg#name
# gf:name rdf:type owl:ObjectProperty ;
#         rdfs:domain gf:Person ;
#         rdfs:range xsd:Xsdstring ;
#         rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#         rdfs:label "name" .

g.add((FOAF.name, RDFS.domain, FOAF.Person))
g.add((FOAF.name, RDFS.range, XSD.string))
g.add((FOAF.name, RDFS.label, Literal("name")))



###  http://www.gra.fo/schema/untitled-ekg#of_type
# gf:of_type rdf:type owl:ObjectProperty ;
#            rdfs:domain gf:Submission ;
#            rdfs:range gf:PaperType ;
#            rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#            rdfs:label "of_type" .

g.add((KG_SDM.of_type, RDFS.domain, KG_SDM.Submission))
g.add((KG_SDM.of_type, RDFS.range, KG_SDM.PaperType))
g.add((KG_SDM.of_type, RDFS.label, Literal("of_type")))








###  http://www.gra.fo/schema/untitled-ekg#paper_title
# gf:paper_title rdf:type owl:ObjectProperty ;
#                rdfs:domain gf:Submission ;
#                rdfs:range xsd:Xsdstring_664ad7aa-201d-4cb4-9b5f-c83e573db337 ;
#                rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#                rdfs:label "paper_title" .

g.add((KG_SDM.paper_title, RDFS.domain, KG_SDM.Submission))
g.add((KG_SDM.paper_title, RDFS.range, XSD.string))
g.add((KG_SDM.paper_title, RDFS.label, Literal("paper_title")))







# ###  http://www.gra.fo/schema/untitled-ekg#paper_year
# gf:paper_year rdf:type owl:ObjectProperty ;
#               rdfs:domain gf:Submission ;
#               rdfs:range xsd:Xsdinteger ;
#               rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#               rdfs:label "paper_year" .


g.add((KG_SDM.paper_year, RDFS.domain, KG_SDM.Submission))
g.add((KG_SDM.paper_year, RDFS.range, XSD.integer))
g.add((KG_SDM.paper_year, RDFS.label, Literal("paper_year")))








###  http://www.gra.fo/schema/untitled-ekg#publication_title
# gf:publication_title rdf:type owl:ObjectProperty ;
#                      rdfs:domain gf:Publication ;
#                      rdfs:range xsd:Xsdstring_a8c526b2-2486-4f15-827b-608dc472e788 ;
#                      rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#                      rdfs:label "publication_title" .


g.add((KG_SDM.publication_title, RDFS.domain, KG_SDM.Publication))
g.add((KG_SDM.publication_title, RDFS.range, XSD.string))
g.add((KG_SDM.publication_title, RDFS.label, Literal("publication_title")))




# ###  http://www.gra.fo/schema/untitled-ekg#publication_year
# gf:publication_year rdf:type owl:ObjectProperty ;
#                     rdfs:domain gf:Publication ;
#                     rdfs:range xsd:Xsdinteger_2c406144-6ecf-41d6-b096-f63a4f137f1b ;
#                     rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#                     rdfs:label "publication_year" .

g.add((KG_SDM.publication_year, RDFS.domain, KG_SDM.Publication))
g.add((KG_SDM.publication_year, RDFS.range, XSD.integer))
g.add((KG_SDM.publication_year, RDFS.label, Literal("publication_year")))





# ###  http://www.gra.fo/schema/untitled-ekg#published_in
# gf:published_in rdf:type owl:ObjectProperty ;
#                 rdfs:domain gf:Submission ;
#                 rdfs:range gf:Publication ;
#                 rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#                 rdfs:label "published_in" .

g.add((KG_SDM.published_in, RDFS.domain, KG_SDM.Submission))
g.add((KG_SDM.published_in, RDFS.range, KG_SDM.Publication))
g.add((KG_SDM.published_in, RDFS.label, Literal("published_in")))



# ###  http://www.gra.fo/schema/untitled-ekg#related_to
# gf:related_to rdf:type owl:ObjectProperty ;
#               rdfs:domain gf:Submission ;
#               rdfs:range gf:Keyword ;
#               rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#               rdfs:label "related_to" .

g.add((KG_SDM.related_to, RDFS.domain, KG_SDM.Submission))
g.add((KG_SDM.related_to, RDFS.range, KG_SDM.Keyword))
g.add((KG_SDM.related_to, RDFS.label, Literal("related_to")))




# ###  http://www.gra.fo/schema/untitled-ekg#reviews
# gf:reviews rdf:type owl:ObjectProperty ;
#            rdfs:domain gf:Reviewers ;
#            rdfs:range gf:Review_process ;
#            rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#            rdfs:label "reviews" .


g.add((KG_SDM.reviews, RDFS.domain, KG_SDM.Reviewers))
g.add((KG_SDM.reviews, RDFS.range, KG_SDM.Review_process))
g.add((KG_SDM.reviews, RDFS.label, Literal("reviews")))




# ###  http://www.gra.fo/schema/untitled-ekg#school
# gf:school rdf:type owl:ObjectProperty ;
#           rdfs:domain gf:Academic ;
#           rdfs:range xsd:Xsdstring_fb6d261b-f832-467d-9209-74cf6b8208fa ;
#           rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#           rdfs:label "school" .

g.add((KG_SDM.school, RDFS.domain, KG_SDM.Academic))
g.add((KG_SDM.school, RDFS.range, XSD.string))
g.add((KG_SDM.school, RDFS.label, Literal("school")))



# ###  http://www.gra.fo/schema/untitled-ekg#submitted_to
# gf:submitted_to rdf:type owl:ObjectProperty ;
#                 rdfs:domain gf:Submission ;
#                 rdfs:range gf:Venue ;
#                 rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#                 rdfs:label "submitted_to" .

g.add((KG_SDM.submitted_to, RDFS.domain, KG_SDM.Submission))
g.add((KG_SDM.submitted_to, RDFS.range, KG_SDM.Venue))
g.add((KG_SDM.submitted_to, RDFS.label, Literal("submitted_to")))






# ###  http://www.gra.fo/schema/untitled-ekg#venue_title
# gf:venue_title rdf:type owl:ObjectProperty ;
#                rdfs:domain gf:Venue ;
#                rdfs:range gf:Xsdstring ;
#                rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#                rdfs:label "venue_title" .

g.add((KG_SDM.venue_title, RDFS.domain, KG_SDM.Venue))
g.add((KG_SDM.venue_title, RDFS.range, XSD.string))
g.add((KG_SDM.venue_title, RDFS.label, Literal("venue_title")))




# ###  http://www.gra.fo/schema/untitled-ekg#writes
# gf:writes rdf:type owl:ObjectProperty ;
#           rdfs:domain gf:Author_6b6d5199-b3e0-474f-a06c-1353577dcb8c ;
#           rdfs:range gf:Submission ;
#           rdfs:isDefinedBy <http://www.gra.fo/schema/untitled-ekg> ;
#           rdfs:label "writes" .

g.add((KG_SDM.writes, RDFS.domain, KG_SDM.Author))
g.add((KG_SDM.writes, RDFS.range, KG_SDM.Submission))
g.add((KG_SDM.writes, RDFS.label, Literal("writes")))

```
